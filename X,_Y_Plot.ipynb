```python
import matplotlib.pyplot as plt
import numpy as np 

plt.style.use('seaborn-dark')
# Make data
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x) 

# Plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
